## **Carga de datos**
Los datos se obtienen a partir de dos archivos **CSV** delimitados por punto y coma (;) importanto la librería csv:<br>
```
archivo_natalidad = 'sample_data/nacimientos.csv'
archivo_expectativa = 'sample_data/expectativa.csv'
```
Ambos archivos cuentan con columnas que no son significativas para nuestro análisis, por ejemplo: barrio y códigos. Entonces, como parte de la **limpieza de datos**, las descartamos al principio para confeccionar los diccionarios<sup>1</sup> solo con los datos más relevantes.<br>
Finalmente, la estructura de los <b>diccionarios</b> elegida, luego del proceso identificación de variables para nuestro trabajo, es la siguiente:<br><br>
**dic_natalidad**
```
{1: {'distrito':'', 'genero':'', 'nacimientos': 0 }}
```

**dic_expectativa**
```
{1: {'distrito':'', 'genero':'', 'expectativa': 0 }}
```
**Nota**<sup>1</sup>: utilizamos diccionarios dado que su estructura nos permite almacenar distintos tipos de valores (str, int, etc), identificar elementos por una clave y modificar su longitud agregando o sacando elementos.

````python
#librerías importadas
import csv
import numpy as np
import matplotlib.pyplot as plt

#archivos de trabajo
archivo_natalidad = 'sample_data/nacimientos.csv'
archivo_expectativa = 'sample_data/expectativa.csv'

#diccionarios para almacenar los datos
dic_natalidad = {}
dic_expectativa = {}


#función de lectura del archivo de datos de natalidad
#almacena en el diccionario dic_natalidad
def carga_natalidad():
  with open(archivo_natalidad, mode='r') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=';')
    n = 0
    for row in csv_reader:
      #salteo la primera fila, de encabezados, guardo el resto
      if n > 0:
        #agregado del item al diccionario, sólo teniendo en cuenta las variables relevantes
        dic_natalidad[n] = {'distrito': row[2], 'genero': row[5], 'nacimientos': int(row[6])}
      n += 1

#función de lectura del archivo de datos de expectativa
#almacena en el diccionario dic_expectativa
def carga_expectativa():
  with open(archivo_expectativa, mode='r') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=';')
    n = 0 #contador como clave
    for row in csv_reader:
      #salteo la primera fila, de encabezados, guardo el resto
      if n > 0:
        #la expectativa esta registrada en cinco períodos
        #se calcula el promedio
        #los períodos se agregan a la lista valores
        valores = []
        for i in range(2,7):
          if row[i] != '':
            valores.append(float(row[i])) 
        #si la lista tiene datos, se agrega el item al diccionario, calculando el promedio de la lista
        if len(valores)>0:      
          dic_expectativa[n] = {'distrito': row[0],'genero': row[7], 'expectativa': np.average(valores)}
      n += 1


#ejecución de la carga de los diccionarios -> llamo a la función
#natalidad
carga_natalidad()
print('Diccionario con datos de natalidad\n')
print(f'\t{dic_natalidad}')
print(f'\t{len(dic_natalidad.keys())} registros\n')

#expectativa de vida
carga_expectativa()
print('Diccionario con datos de expectativa de vida\n')
print(f'\t{dic_expectativa}')
print(f'\t{len(dic_expectativa.keys())} registros\n')
````

## **Preparación de datos**: 

Nuestro análisis se compone de los diez distritos de Cataluña, para agilizar el proceso del cálculo de las variables, armamos una **lista** con todos estos para usar en ambos diccionarios:<br>
```
distritos
```
Asimismo, los datos de expectativa de vida están organizados en cinco intervalos de cinco años cada uno (2006-2010, 2007-2011, 2008-2012, 2009-2013, 2010-2014). Mientras que, los datos de natalidad están organizados por año desde 2013 a 2017. Por esta razón, para proceder en un análisis global, decidimos utilizar el siguiente criterio: 
* **Expectativa de vida**: promedio de todos los períodos.
* **Natalidad**: cantidad acumulada de todos los años.

Ahora bien, para simplificar los cálculos y el desarrollo de las visualizaciones creamos un **diccionario integral** que incluye los valores totales por distritos de los diccionarios anteriormente mencionados:
```
datos = {'distrito': {'exp_f': 0, 'exp_m': 0, 'nat_f': 0, 'nat_m': 0, 'nat_total': 0}}
```
La clave es el nombre del distrito y los valores son:<br>

* exp_f: expectativa de vida femenina
* exp_m: expectativa de vida masculina
* nat_f: natalidad mujeres
* nat_m: natalidad varones
* nat_total: total de nacimientos


````python
#lista de distritos
distritos = []
#recorrida del diccionario de natalidad para obtener distritos
for v in dic_natalidad.values():
  #en este diccionario hay registros con distrito indefinido, denominado "No consta". No se agrega al diccionario
  #por lo tanto sus valores no se tienen en cuenta en los análisis
  if v.get('distrito') not in distritos and v.get('distrito') != 'No consta':
    distritos.append(v.get('distrito'))

#recorrida del diccionario de expectativa de vida para obtener distritos
for v in dic_expectativa.values():
  if v.get('distrito') not in distritos :
    distritos.append(v.get('distrito'))

#se ordenan alfabéticamente
distritos.sort()

print('Distritos\n')
print(f'\t{distritos}')
print(f'\t{len(distritos)} distritos')
````

````python
#diccionario de datos para trabajar
datos = {}
#para cada distrito
for d in distritos:
  #preparación del item de diccionario -> armo la estrutura 
  #d es la clave de datos
  datos[d] = {'exp_f': 0, 'exp_m': 0, 'nat_f': 0, 'nat_m': 0, 'nat_total': 0}

  #datos de expectativa de vida
  #se calcula el promedio agregando para el distrito el valor por género
  #en dos listas
  valores_f = []
  valores_m = []
  for v in dic_expectativa.values():
    if v.get('distrito') == d:
      if v.get('genero') == 'Female':
        valores_f.append(v.get('expectativa')) #expectativa es la clave de dic_expectativas
      else:
        valores_m.append(v.get('expectativa'))
  #expectativa por género
  if len(valores_f) > 0:
    datos[d]['exp_f'] = round(np.average(valores_f),2)
  if len(valores_f) > 0:
    datos[d]['exp_m'] = round(np.average(valores_m),2)

  #datos de natalidad
  #se calcula el total de nacimientos por género para el distrito
  valores_f = []
  valores_m = []
  for v in dic_natalidad.values():
    if v.get('distrito') == d:
      if v.get('genero') == 'Girls':
        valores_f.append(v.get('nacimientos'))
      else:
        valores_m.append(v.get('nacimientos'))
  #natalidad por género
  if len(valores_f) > 0:
    datos[d]['nat_f'] = np.sum(valores_f)
  if len(valores_f) > 0:
    datos[d]['nat_m'] = np.sum(valores_m)
  #natalidad total
  datos[d]['nat_total'] = datos[d]['nat_f'] + datos[d]['nat_m']

print('Diccionario de datos\n')
print(f'\t{datos}')
print(f'\t{len(datos)} registros')

````

## **Análisis**
En esta sección se desarrolla el cálculo de los indicadores solicitados respondiento tres consignas.

# **Consigna # 1**
Para cada distrito obtener el promedio de expectativa de vida, discriminado por genero.

````python
print("(Consigna: Para cada distrito obtener el promedio de expectativa de vida, discriminado por genero.)\n\n")

# Iteramos por los items de datos, y guardamos en variables el dato de expectativa de vida discriminado por genero
for k, v in datos.items():
  exp_m = v["exp_m"]
  exp_f = v["exp_f"]

  # Luego utilizamos las variables en este print
  print(f"'{k}' tiene una expectativa de vida femenina de ({exp_f}) años, y masculina de ({exp_m}) años")
````

# **Consigna # 2**
Obtener el/los distritos con mayor y menor natalidad respectivamente.

````python
print("(Consigna: Obtener el/los distritos con mayor y menor natalidad respectivamente)\n\n")

# Creamos listas que seran usadas más adelante
nacimientos = {}
distritos_menor_natalidad = {}
distritos_mayor_natalidad = {}

for k, v in datos.items():

  # Modificamos el diccionario nacimientos, para que quede compuesta con el formato de (key = distrito) : (value = natalidad)
  nacimientos[k] = v["nat_m"] + v["nat_f"]

  # Creamos variable conteniendo a los values del diccionario nacimientos, ordenado de menor a mayor
  valores_ordenado = sorted(list(nacimientos.values()))

  
# Iteramos los items de nacimientos y modificamos los diccionarios distritos_menor_natalidad y distritos_mayor_natalidad
for k, v in nacimientos.items():

  if v in valores_ordenado[:3]:
    distritos_menor_natalidad[k] = v
  
  if v in valores_ordenado[-3:]:
    distritos_mayor_natalidad[k] = v

# Presentacion de los datos
print("Los distritos con menor natalidad son los siguientes:\n")
for k, v in distritos_menor_natalidad.items():
  print(f"'{k}', con una natalidad de: {v}")

print("\n\nLos distritos con mayor natalidad son los siguientes:\n")
for k, v in distritos_mayor_natalidad.items():
  print(f"'{k}', con una natalidad de: {v}")
````

# **Consigna # 3**
Realizar una comparación entre los distritos con mayor expectativa de vida y su nivel de natalidad


````python
# Creamos variables
exp = {}
distritos_mayor_exp = {}
expectativa_total = []
for k, v in datos.items():
  # Modificamos el diccionario exp para que contenga el promedio general de expectativa de vida
  exp[k] = round((v["exp_m"] + v["exp_f"]) / 2, 2)

for i in exp.values():
  expectativa_total.append(i)

# Creamos lista con los valores de exp
val_ord = sorted(list(exp.values()))

# Guardamos los tres distritos de mayor expectativa en una diccionario
for k, v in exp.items():
  if v in val_ord[-3:]:
    distritos_mayor_exp[k] = v

# Presentacion de los datos
print("(Consigna: Realizar una comparación entre los distritos con mayor expectativa de vida y su nivel de natalidad)\n")
print("Los distritos con mayor expecativa son los siguientes, estan acompañados de su nivel de natalidad.\n")

for k, v in distritos_mayor_exp.items():
  print(f"'{k}' tiene una expectativa de vida de ({v}) años, y un nivel de natalidad de: {nacimientos[k]}")
````
